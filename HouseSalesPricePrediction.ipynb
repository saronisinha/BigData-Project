## Overview

This notebook will show you how to create and query a table or DataFrame that you uploaded to DBFS. [DBFS](https://docs.databricks.com/user-guide/dbfs-databricks-file-system.html) is a Databricks File System that allows you to store data for querying inside of Databricks. This notebook assumes that you have a file already inside of DBFS that you would like to read from.

This notebook is written in **Python** so the default cell type is Python. However, you can use different languages by using the `%LANGUAGE` syntax. Python, Scala, SQL, and R are all supported.

```python
# File location and type
file_location = "/FileStore/tables/HomeSales.csv"
file_type = "csv"

# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(df)
```

#Feature Engineering

```python
from pyspark.sql.functions import *

df = df.withColumn("baths",col("bathsfull")+.5*col("bathshalf"))


from pyspark.sql.functions import when
df=df.withColumn("specialsale", when(df.splsale == 'None', 0)
.otherwise(1))


df=df.withColumn("tileroof", when(df.Roof == 'Title', 1)
.when(df.Roof == 'Slate', 1)
.when(df.Roof == 'Slate,Title', 1)
.when(df.Roof == 'Concrete,Title', 1)
.otherwise(0))



df=df.withColumn("privatepool", when(df.Pool == 'Private', 1)
.when(df.Pool == 'Private,Community', 1)
.otherwise(0))
df=df.withColumn("communitypool", when(df.Pool == 'Community', 1)
.when(df.Pool == 'Private,Community', 1)
.otherwise(0))



spark.sql("set spark.sql.legacy.timeParserPolicy=LEGACY")
df = df.withColumn("Date",to_date(df.PendingDate, 'MM/dd/yyyy'))
df = df.withColumn("year",year(col("Date")))
```

#Feature Processing

```python
# Create a 70-30 train test split

train_data,test_data=df.randomSplit([0.7,0.3])
```

```python
# Import the required libraries

from pyspark.ml.regression import LinearRegression
from pyspark.ml.regression import GeneralizedLinearRegression
from pyspark.ml.feature import VectorAssembler,StringIndexer,StandardScaler
from pyspark.ml import Pipeline
```

```python
# Use StringIndexer to convert the categorical columns to hold numerical data

beds_indexer = StringIndexer(inputCol="Beds",outputCol='beds_index',handleInvalid='keep')
garages_indexer = StringIndexer(inputCol="garages",outputCol="garages_index",handleInvalid='keep')
baths_indexer = StringIndexer(inputCol="baths",outputCol='baths_index',handleInvalid='keep')
year_indexer = StringIndexer(inputCol="year",outputCol="year_index",handleInvalid='keep')
```

```python
assembler = VectorAssembler(inputCols=["sqft","tileroof","yrblt","privatepool","communitypool","specialsale","lppersqft",'beds_index',
                                      'garages_index','baths_index','year_index'],
                           outputCol="features")
```

```python
# Standard scaler is used to scale the data for the linear SVC to perform well on the training data

#scaler = StandardScaler(inputCol="unscaled_features",outputCol="features")
```

```python
pipe = Pipeline(stages=[beds_indexer,garages_indexer,baths_indexer,year_indexer,assembler])
```

##Models

Linear Regression

```python
fitted_pipe=pipe.fit(train_data)
```

```python
train_data_lm=fitted_pipe.transform(train_data)
```

```python
# Create an object for the Linear Regression model

lr_model = LinearRegression(labelCol='pricesold')
```

```python
# Fit the model on the train data

fit_model = lr_model.fit(train_data_lm.select(['features','pricesold']))
```

```python
print("Coefficients: %s" % str(fit_model.coefficients))
print("Intercept: %s" % str(fit_model.intercept))
print(fit_model.summary.pValues)
```

```python
trainingSummary = fit_model.summary
print("numIterations: %d" % trainingSummary.totalIterations)
print("objectiveHistory: %s" % str(trainingSummary.objectiveHistory))
trainingSummary.residuals.show()
print("RMSE: %f" % trainingSummary.rootMeanSquaredError)
print("r2: %f" % trainingSummary.r2)
```

```python
# Transform the test data using the model to predict the duration

test_data_lm=fitted_pipe.transform(test_data)
```

```python
# Store the results in a dataframe
results = fit_model.transform(test_data_lm)
```

```python
#from pyspark.sql.functions import round, col

results_lm = results.select('pricesold', round(col('prediction')))
results_lm = results_lm.withColumnRenamed('round(prediction, 0)','prediction')
results_lm.display()
```

```python
from numpy import polyfit
x1 = results_lm.toPandas()['pricesold'].values.tolist()
y1 = results_lm.toPandas()['prediction'].values.tolist()
plt.scatter(x1, y1, color='red', s=30)
plt.xlabel('Pricesold')
plt.ylabel('Prediction')
plt.title('Linear Regression')
p1 = polyfit(x1, y1, 1)
plt.plot(x1, np.polyval(p1,x1), 'g-' )
plt.show()
```

```python
test_results = fit_model.evaluate(test_data_lm)
```

```python
test_results.residuals.show()
```

```python
test_results.rootMeanSquaredError
```

```python
test_results.r2
```
